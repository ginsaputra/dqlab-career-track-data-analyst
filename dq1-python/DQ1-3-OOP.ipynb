Task: Print the company's total expenses for salaries.

```python
# Define class Employee
class Employee:
    def __init__(self, name, age, income, ot_incentive):
        self.name = name
        self.age = age 
        self.income = income 
        self.additional_pay = 0
        self.ot_incentive = ot_incentive 
    def overtime(self):
        self.additional_pay += self.ot_incentive
    def project_pay(self, total_addition):
        self.additional_pay += total_addition 
    def total_income(self):
        return self.income + self.additional_pay 

# Define child classes of Employee 
class Freelancer(Employee):
    def __init__(self, name, age, income): 
        super().__init__(name, age, income, 0)
    def project_pay(self, project_value): 
        self.additional_pay += project_value * 0.01

class DataAnalyst(Employee):
    def __init__(self, name, age=21, income=6500000, ot_incentive=100000):
        super().__init__(name, age, income, ot_incentive) 

class DataScientist(Employee):
    def __init__(self, name, age=25, income=12000000, ot_incentive=150000):
        super().__init__(name, age, income, ot_incentive) 
    def project_pay(self, project_value):
        self.additional_pay += project_value * 0.1

# Define child classes of Freelancer
class DataCleaner(Freelancer):
    def __init__(self, name, age, income = 4000000): 
        super().__init__(name, age, income)

class TechnicalDocumenter(Freelancer):
    def __init__(self, name, age, income = 2500000): 
        super().__init__(name, age, income)
        def project_pay(self,  total_addition): 
            return

# Define class Company 
class Company:
    def __init__(self, name, address, phone_number): 
        self.name = name
        self.address = address 
        self.phone_number = phone_number 
        self.list_employee = []

    def activate_employee(self, employee): 
        self.list_employee.append(employee)

    def nonactivate_employee(self, employee_name): 
        nonactive_employee = None
        for employee in self.list_employee:
            if employee.name == employee_name: 
                nonactive_employee = employee 
                break
        if nonactive_employee is not None: 
            self.list_employee.remove(nonactive_employee)
            
    def total_expenses(self): 
        expenses = 0
        for employee in self.list_employee:
            expenses += employee.total_income() 
        return expenses

    def find_employee(self, employee_name): 
        for employee in self.list_employee:
            if employee.name == employee_name: 
                return employee
        return None

# Create Employee objects based on the table above
ani = DataCleaner('Ani', 25)
budi = TechnicalDocumenter('Budi', 18)
cici = DataScientist('Cici')
didi = DataScientist('Didi', 32, 20000000)
efi = DataAnalyst('Efi')
febi = DataAnalyst('Febi', 28, 12000000)

# Create Company object
Company = Company('ABC', 'Jl. Jendral Sudirman, Blok 11', '(021) 95812XX')

# Activate each employee
Company.activate_employee(ani)
Company.activate_employee(budi)
Company.activate_employee(cici)
Company.activate_employee(didi)
Company.activate_employee(efi)
Company.activate_employee(febi)

print(f"Company's total expenses for salary: Rp{Company.total_expenses()}")
```
